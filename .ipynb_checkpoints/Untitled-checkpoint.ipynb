```python
# =============================================================================
# CELL 3: Enhanced Document Processing (Same as before)
# =============================================================================

class DocumentProcessor:
    """Enhanced document parsing with better error handling"""
    
    @staticmethod
    def extract_text_from_pdf(file_content: bytes) -> str:
        """Extract text from PDF using multiple methods"""
        methods_tried = []
        
        try:
            # Method 1: PyPDF2
            pdf_file = BytesIO(file_content)
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
            methods_tried.append("PyPDF2")
            
            if len(text.strip()) > 50:
                print(f"✅ PDF extracted using PyPDF2 ({len(text)} chars)")
                return text
        except Exception as e:
            print(f"❌ PyPDF2 failed: {e}")
        
        try:
            # Method 2: pdfplumber (better for complex layouts)
            pdf_file = BytesIO(file_content)
            with pdfplumber.open(pdf_file) as pdf:
                text = ""
                for page in pdf.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + "\n"
            methods_tried.append("pdfplumber")
            
            if len(text.strip()) > 50:
                print(f"✅ PDF extracted using pdfplumber ({len(text)} chars)")
                return text
        except Exception as e:
            print(f"❌ pdfplumber failed: {e}")
        
        raise Exception(f"Could not extract text from PDF. Tried: {', '.join(methods_tried)}")
    
    @staticmethod
    def extract_text_from_docx(file_content: bytes) -> str:
        """Extract text from DOCX with better handling"""
        try:
            doc = docx.Document(BytesIO(file_content))
            text = ""
            for paragraph in doc.paragraphs:
                text += paragraph.text + "\n"
            
            # Also extract text from tables
            for table in doc.tables:
                for row in table.rows:
                    for cell in row.cells:
                        text += cell.text + " "
                text += "\n"
            
            print(f"✅ DOCX extracted ({len(text)} chars)")
            return text
        except Exception as e:
            print(f"❌ DOCX extraction failed: {e}")
            return ""
    
    @staticmethod
    def extract_text_from_txt(file_content: bytes) -> str:
        """Extract text from TXT with encoding detection"""
        encodings = ['utf-8', 'latin-1', 'cp1252', 'iso-8859-1']
        
        for encoding in encodings:
            try:
                text = file_content.decode(encoding)
                print(f"✅ TXT extracted using {encoding} ({len(text)} chars)")
                return text
            except UnicodeDecodeError:
                continue
        
        raise Exception("Could not decode text file with any common encoding")
    
    @classmethod
    def process_resume(cls, file_content: bytes, file_extension: str) -> str:
        """Main method with enhanced error handling"""
        file_extension = file_extension.lower().strip()
        
        if file_extension == '.pdf':
            return cls.extract_text_from_pdf(file_content)
        elif file_extension in ['.docx', '.doc']:
            return cls.extract_text_from_docx(file_content)
        elif file_extension == '.txt':
            return cls.extract_text_from_txt(file_content)
        else:
            raise ValueError(f"Unsupported file format: {file_extension}")

print("📄 Enhanced document processor ready!")
```

```python
# =============================================================================
# CELL 4: FREE NLP Processing with Enhanced Skills Database
# =============================================================================

class FreeNLPProcessor:
    """Enhanced NLP processor using only free tools"""
    
    def __init__(self):
        # Comprehensive skills database (can be loaded from JSON file)
        self.technical_skills = {
            'programming_languages': [
                'python', 'java', 'javascript', 'c++', 'c#', 'ruby', 'php', 'go', 
                'rust', 'kotlin', 'swift', 'typescript', 'scala', 'r', 'matlab',
                'perl', 'dart', 'elixir', 'haskell', 'clojure', 'groovy', 'lua'
            ],
            'web_technologies': [
                'html', 'css', 'react', 'angular', 'vue', 'nodejs', 'express',
                'django', 'flask', 'spring', 'bootstrap', 'tailwind', 'sass',
                'webpack', 'babel', 'nextjs', 'nuxtjs', 'svelte', 'jquery'
            ],
            'databases': [
                'sql', 'mysql', 'postgresql', 'mongodb', 'redis', 'elasticsearch',
                'sqlite', 'oracle', 'cassandra', 'dynamodb', 'neo4j', 'couchdb',
                'influxdb', 'mariadb', 'firebase', 'supabase'
            ],
            'cloud_platforms': [
                'aws', 'azure', 'gcp', 'docker', 'kubernetes', 'terraform',
                'jenkins', 'git', 'github', 'gitlab', 'bitbucket', 'heroku',
                'vercel', 'netlify', 'digitalocean', 'linode'
            ],
            'data_science': [
                'pandas', 'numpy', 'scikit-learn', 'tensorflow', 'pytorch',
                'matplotlib', 'seaborn', 'jupyter', 'spark', 'hadoop',
                'tableau', 'powerbi', 'plotly', 'statsmodels', 'keras'
            ],
            'mobile_development': [
                'ios', 'android', 'react native', 'flutter', 'xamarin',
                'ionic', 'cordova', 'swift', 'objective-c', 'kotlin'
            ],
            'devops_tools': [
                'docker', 'kubernetes', 'jenkins', 'gitlab ci', 'github actions',
                'ansible', 'puppet', 'chef', 'nagios', 'prometheus', 'grafana'
            ],
            'soft_skills': [
                'leadership', 'communication', 'teamwork', 'problem-solving',
                'project management', 'agile', 'scrum', 'critical thinking',
                'time management', 'collaboration', 'adaptability'
            ]
        }
        
        # Flatten for easier searching
        self.all_skills = []
        for category, skills in self.technical_skills.items():
            self.all_skills.extend(skills)
        
        # Common job role patterns
        self.job_roles = [
            'software engineer', 'data scientist', 'product manager', 'designer',
            'frontend developer', 'backend developer', 'full stack', 'devops',
            'machine learning engineer', 'data analyst', 'product owner'
        ]
    
    def preprocess_text(self, text: str) -> str:
        """Enhanced text preprocessing"""
        if not text:
            return ""
        
        # Basic cleaning
        text = re.sub(r'\s+', ' ', text)  # Normalize whitespace
        text = text.strip()
        
        # Remove email addresses and phone numbers (privacy)
        text = re.sub(r'\S+@\S+\.\S+', '[EMAIL]', text)
        text = re.sub(r'(\+?\d{1,3}[-.\s]?)?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}', '[PHONE]', text)
        
        # Normalize common variations
        text = re.sub(r'\bjavascript\b', 'javascript', text, flags=re.IGNORECASE)
        text = re.sub(r'\bnode\.?js\b', 'nodejs', text, flags=re.IGNORECASE)
        text = re.sub(r'\breact\.?js\b', 'react', text, flags=re.IGNORECASE)
        
        return text
    
    def extract_skills_advanced(self, text: str) -> Dict[str, List[str]]:
        """Advanced skill extraction using multiple methods"""
        text_lower = text.lower()
        found_skills = {category: [] for category in self.technical_skills.keys()}
        
        # Method 1: Direct keyword matching with context
        for category, skills in self.technical_skills.items():
            for skill in skills:
                # Use word boundaries and context
                patterns = [
                    rf'\b{re.escape(skill)}\b',
                    rf'\b{re.escape(skill)}\.js\b',
                    rf'\b{re.escape(skill)}\s*(programming|development|experience)\b'
                ]
                
                for pattern in patterns:
                    if re.search(pattern, text_lower):
                        if skill not in found_skills[category]:
                            found_skills[category].append(skill)
                        break
        
        # Method 2: Use spaCy NER for additional entities
        doc = nlp(text)
        tech_entities = []
        for ent in doc.ents:
            if ent.label_ in ['ORG', 'PRODUCT', 'LANGUAGE', 'GPE']:
                entity_text = ent.text.lower().strip()
                if len(entity_text) > 2 and not entity_text.isdigit():
                    tech_entities.append(entity_text)
        
        # Cross-reference entities with skills
        for entity in tech_entities:
            for category, skills in self.technical_skills.items():
                for skill in skills:
                    if skill in entity or entity in skill:
                        if skill not in found_skills[category]:
                            found_skills[category].append(skill)
        
        # Method 3: Extract years of experience
        experience_patterns = [
            r'(\d+)\+?\s*years?\s*(?:of\s*)?experience',
            r'(\d+)\+?\s*yrs?\s*(?:of\s*)?experience',
            r'experience.*?(\d+)\+?\s*years?'
        ]
        
        experience_years = []
        for pattern in experience_patterns:
            matches = re.findall(pattern, text_lower)
            for match in matches:
                try:
                    years = int(match)
                    if 0 < years <= 50:  # Reasonable range
                        experience_years.append(years)
                except:
                    continue
        
        # Add experience level to metadata
        if experience_years:
            avg_experience = sum(experience_years) / len(experience_years)
            found_skills['experience_level'] = [f"{avg_experience:.1f} years"]
        
        return found_skills
    
    def extract_key_phrases_enhanced(self, text: str, n_phrases: int = 15) -> List[Dict]:
        """Enhanced key phrase extraction with scores"""
        if not text or len(text.strip()) < 50:
            return []
        
        try:
            # Split into sentences
            sentences = [s.strip() for s in text.split('.') if len(s.strip()) > 10]
            
            if len(sentences) < 2:
                return []
            
            # TF-IDF extraction
            vectorizer = TfidfVectorizer(
                max_features=200,
                stop_words='english',
                ngram_range=(1, 3),
                min_df=1,
                max_df=0.8
            )
            
            tfidf_matrix = vectorizer.fit_transform(sentences)
            feature_names = vectorizer.get_feature_names_out()
            scores = tfidf_matrix.sum(axis=0).A1
            
            # Get top phrases with scores
            phrase_scores = list(zip(feature_names, scores))
            phrase_scores.sort(key=lambda x: x[1], reverse=True)
            
            # Return top phrases with metadata
            key_phrases = []
            for phrase, score in phrase_scores[:n_phrases]:
                key_phrases.append({
                    'phrase': phrase,
                    'score': round(float(score), 3),
                    'category': self._categorize_phrase(phrase)
                })
            
            return key_phrases
            
        except Exception as e:
            print(f"Key phrase extraction failed: {e}")
            return []
    
    def _categorize_phrase(self, phrase: str) -> str:
        """Categorize phrases into types"""
        phrase_lower = phrase.lower()
        
        # Check against skill categories
        for category, skills in self.technical_skills.items():
            for skill in skills:
                if skill in phrase_lower:
                    return category.replace('_', ' ').title()
        
        # Check for action words
        action_words = ['developed', 'implemented', 'managed', 'led', 'created', 'designed']
        if any(word in phrase_lower for word in action_words):
            return 'Achievement'
        
        # Check for metrics
        if re.search(r'\d+%|\$\d+|\d+\s*(years|months)', phrase_lower):
            return 'Quantifiable'
        
        return 'General'
    
    def analyze_text_quality(self, text: str) -> Dict:
        """Analyze resume text quality and provide insights"""
        analysis = {
            'word_count': len(text.split()),
            'character_count': len(text),
            'sentence_count': len([s for s in text.split('.') if s.strip()]),
            'avg_sentence_length': 0,
            'has_quantifiable_achievements': False,
            'has_action_verbs': False,
            'quality_score': 0
        }
        
        sentences = [s.strip() for s in text.split('.') if s.strip()]
        if sentences:
            analysis['avg_sentence_length'] = sum(len(s.split()) for s in sentences) / len(sentences)
        
        # Check for quantifiable achievements
        metrics_pattern = r'\d+%|\$\d+|\d+\s*(million|thousand|k|years|months|projects|teams)'
        analysis['has_quantifiable_achievements'] = bool(re.search(metrics_pattern, text, re.IGNORECASE))
        
        # Check for action verbs
        action_verbs = ['achieved', 'implemented', 'developed', 'managed', 'led', 'created', 
                       'improved', 'increased', 'reduced', 'optimized', 'designed']
        analysis['has_action_verbs'] = any(verb in text.lower() for verb in action_verbs)
        
        # Calculate quality score
        score = 0
        if analysis['word_count'] > 200: score += 20
        if analysis['has_quantifiable_achievements']: score += 25
        if analysis['has_action_verbs']: score += 25
        if 10 < analysis['avg_sentence_length'] < 25: score += 15
        if analysis['sentence_count'] > 10: score += 15
        
        analysis['quality_score'] = score
        
        return analysis

print("🧠 Enhanced FREE NLP processor ready!")
```

```python
# =============================================================================
# CELL 5: FREE Resume-Job Matcher (No APIs Required)
# =============================================================================

class FreeResumeJobMatcher:
    """FREE matching engine using only open-source models"""
    
    def __init__(self):
        self.nlp_processor = FreeNLPProcessor()
    
    def generate_embeddings(self, texts: List[str]) -> np.ndarray:
        """Generate embeddings using free SentenceTransformers"""
        print(f"🔄 Generating embeddings for {len(texts)} texts...")
        embeddings = sentence_model.encode(texts, convert_to_tensor=False, show_progress_bar=True)
        print(f"✅ Generated embeddings shape: {embeddings.shape}")
        return embeddings
    
    def calculate_similarity_advanced(self, resume_text: str, job_description: str) -> Dict:
        """Advanced similarity calculation with multiple metrics"""
        
        # 1. Semantic similarity using embeddings
        texts = [resume_text, job_description]
        embeddings = self.generate_embeddings(texts)
        semantic_similarity = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
        
        # 2. Keyword overlap similarity
        resume_words = set(resume_text.lower().split())
        job_words = set(job_description.lower().split())
        keyword_overlap = len(resume_words.intersection(job_words)) / len(job_words.union(resume_words))
        
        # 3. TF-IDF similarity
        try:
            vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
            tfidf_matrix = vectorizer.fit_transform([resume_text, job_description])
            tfidf_similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
        except:
            tfidf_similarity = 0.0
        
        return {
            'semantic_similarity': float(semantic_similarity),
            'keyword_overlap': float(keyword_overlap),
            'tfidf_similarity': float(tfidf_similarity),
            'overall_similarity': float((semantic_similarity + keyword_overlap + tfidf_similarity) / 3)
        }
    
    def analyze_skill_gaps_detailed(self, resume_skills: Dict, job_skills: Dict) -> Dict:
        """Detailed skill gap analysis with recommendations"""
        gaps = {
            'missing_skills': {},
            'matching_skills': {},
            'skill_coverage': {},
            'priority_gaps': {},
            'strength_areas': {}
        }
        
        total_missing = 0
        total_matching = 0
        
        for category in resume_skills.keys():
            if category == 'experience_level':
                continue
                
            resume_set = set(resume_skills[category])
            job_set = set(job_skills[category])
            
            missing = job_set - resume_set
            matching = job_set.intersection(resume_set)
            
            gaps['missing_skills'][category] = list(missing)
            gaps['matching_skills'][category] = list(matching)
            
            # Calculate coverage
            if len(job_set) > 0:
                coverage = len(matching) / len(job_set) * 100
                gaps['skill_coverage'][category] = round(coverage, 2)
                
                # Identify priority gaps (categories with low coverage)
                if coverage < 50 and len(missing) > 0:
                    gaps['priority_gaps'][category] = list(missing)
                
                # Identify strength areas (high coverage)
                if coverage > 80:
                    gaps['strength_areas'][category] = list(matching)
            else:
                gaps['skill_coverage'][category] = 100.0
            
            total_missing += len(missing)
            total_matching += len(matching)
        
        # Overall statistics
        gaps['total_skills_required'] = total_missing + total_matching
        gaps['total_skills_matched'] = total_matching
        gaps['overall_skill_coverage'] = (total_matching / (total_missing + total_matching) * 100) if (total_missing + total_matching) > 0 else 100
        
        return gaps
    
    def generate_comprehensive_report(self, resume_text: str, job_description: str) -> Dict:
        """Generate comprehensive FREE analysis report"""
        print("📊 Generating comprehensive analysis report...")
        
        # 1. Extract skills from both texts
        print("🔍 Extracting skills...")
        resume_skills = self.nlp_processor.extract_skills_advanced(resume_text)
        job_skills = self.nlp_processor.extract_skills_advanced(job_description)
        
        # 2. Calculate multiple similarity metrics
        print("📈 Calculating similarities...")
        similarity_metrics = self.calculate_similarity_advanced(resume_text, job_description)
        
        # 3. Detailed skill gap analysis
        print("🎯 Analyzing skill gaps...")
        skill_gaps = self.analyze_skill_gaps_detailed(resume_skills, job_skills)
        
        # 4. Extract key phrases
        print("🔑 Extracting key phrases...")
        resume_phrases = self.nlp_processor.extract_key_phrases_enhanced(resume_text)
        job_phrases = self.nlp_processor.extract_key_phrases_enhanced(job_description)
        
        # 5. Analyze text quality
        print("✅ Analyzing text quality...")
        text_quality = self.nlp_processor.analyze_text_quality(resume_text)
        
        # 6. Calculate weighted match score
        weights = {
            'semantic': 0.4,
            'keyword': 0.3,
            'tfidf': 0.2,
            'skill_coverage': 0.1
        }
        
        match_score = (
            similarity_metrics['semantic_similarity'] * weights['semantic'] +
            similarity_metrics['keyword_overlap'] * weights['keyword'] +
            similarity_metrics['tfidf_similarity'] * weights['tfidf'] +
            skill_gaps['overall_skill_coverage']/100 * weights['skill_coverage']
        ) * 100
        
        # 7. Determine match level
        if match_score >= FreeConfig.HIGH_MATCH_THRESHOLD * 100:
            match_level = "High"
            match_color = "green"
        elif match_score >= FreeConfig.MEDIUM_MATCH_THRESHOLD * 100:
            match_level = "Medium"
            match_color = "orange"
        else:
            match_level = "Low"
            match_color = "red"
        
        print(f"✅ Analysis complete! Match score: {match_score:.1f}%")
        
        return {
            'match_score': round(match_score, 2),
            'match_level': match_level,
            'match_color': match_color,
            'similarity_metrics': similarity_metrics,
            'resume_skills': resume_skills,
            'job_skills': job_skills,
            'skill_gaps': skill_gaps,
            'resume_key_phrases': resume_phrases,
            'job_key_phrases': job_phrases,
            'text_quality_analysis': text_quality,
            'analysis_metadata': {
                'total_resume_words': len(resume_text.split()),
                'total_job_words': len(job_description.split()),
                'processing_method': 'free_models_only'
            }
        }

print("🎯 FREE resume matcher ready!")
```

```python
# =============================================================================
# CELL 6: FREE AI Enhancement (Template-Based, No APIs)
# =============================================================================

class FreeAIEnhancer:
    """FREE AI enhancement using templates and rules - NO APIs required!"""
    
    def __init__(self):
        self.config = FreeConfig()
    
    def generate_free_improvements(self, resume_text: str, job_description: str, 
                                 skill_gaps: Dict, text_quality: Dict) -> List[Dict]:
        """Generate improvement suggestions using FREE templates"""
        
        suggestions = []
        
        # 1. Keyword-based suggestions
        missing_skills_flat = []
        for category, skills in skill_gaps['missing_skills'].items():
            missing_skills_flat.extend(skills[:3])  # Top 3 from each category
        
        if missing_skills_flat:
            suggestion = random.choice(self.config.IMPROVEMENT_TEMPLATES['keywords'])
            suggestions.append({
                'type': 'Keywords',
                'suggestion': f"{suggestion}. Focus on: {', '.join(missing_skills_flat[:5])}",
                'impact': f"Could improve keyword matching by addressing {len(missing_skills_flat)} missing skills",
                'priority': 'High' if len(missing_skills_flat) > 5 else 'Medium'
            })
        
        # 2. Quantification suggestions
        if not text_quality.get('has_quantifiable_achievements', False):
            suggestion = random.choice(self.config.IMPROVEMENT_TEMPLATES['quantification'])
            suggestions.append({
                'type': 'Quantification',
                'suggestion': suggestion,
                'impact': "Quantified achievements are 40% more likely to get noticed by recruiters",
                'priority': 'High'
            })
        
        # 3. Skills reorganization
        if skill_gaps.get('priority_gaps'):
            priority_categories = list(skill_gaps['priority_gaps'].keys())
            suggestion = random.choice(self.config.IMPROVEMENT_TEMPLATES['skills'])
            suggestions.append({
                'type': 'Skills',
                'suggestion': f"{suggestion}. Prioritize: {', '.join(priority_categories[:2])}",
                'impact': f"Addresses critical skill gaps in {len(priority_categories)} categories",
                'priority': 'High'
            })
        
        # 4. Content structure
        if text_quality.get('avg_sentence_length', 0) > 25:
            suggestions.append({
                'type': 'Format',
                'suggestion': "Break down long sentences into shorter, more impactful statements",
                'impact': "Improves readability and ATS parsing accuracy",
                'priority': 'Medium'
            })
        
        # 5. Action verb enhancement
        if not text_quality.get('has_action_verbs', False):
            suggestions.append({
                'type': 'Content',
                'suggestion': "Use strong action verbs at the beginning of bullet points (e.g., 'Developed', 'Implemented', 'Led', 'Optimized')",
                'impact': "Action-oriented language increases resume impact by 35%",
                'priority': 'Medium'
            })
        
        # 6. Length optimization
        word_count = text_quality.get('word_count', 0)
        if word_count < 200:
            suggestions.append({
                'type': 'Content',
                'suggestion': "Expand your experience descriptions with more specific details and achievements",
                'impact': "Comprehensive resumes perform better in keyword matching",
                'priority': 'Medium'
            })
        elif word_count > 800:
            suggestions.append({
                'type': 'Format',
                'suggestion': "Consider condensing content to focus on most relevant experiences for this role",
                'impact': "Concise, targeted resumes get more attention from recruiters",
                'priority': 'Low'
            })
        
        # 7. Industry-specific suggestions
        job_lower = job_description.lower()
        if 'senior' in job_lower or 'lead' in job_lower:
            suggestions.append({
                'type': 'Content',
                'suggestion': "Highlight leadership experience, team management, and mentoring responsibilities",
                'impact': "Leadership skills are crucial for senior-level positions",
                'priority': 'High'
            })
        
        # Sort by priority
        priority_order = {'High': 3, 'Medium': 2, 'Low': 1}
        suggestions.sort(key=lambda x: priority_order.get(x['priority'], 0), reverse=True)
        
        return suggestions[:6]  # Return top 6 suggestions
    
    def generate_free_interview_questions(self, job_description: str, resume_text: str, 
                                        resume_skills: Dict) -> List[Dict]:
        """Generate interview questions using FREE templates"""
        
        questions = []
        job_lower = job_description.lower()
        
        # Extract key technologies and skills from job description
        job_skills_flat = []
        for category, skills in resume_skills.items():
            if category != 'experience_level':
                job_skills_flat.extend(skills[:2])  # Top 2 from each category
        
        # 1. Technical Questions
        if job_skills_flat:
            for i in range(min(3, len(job_skills_flat))):
                skill = job_skills_flat[i]
                template = random.choice(self.config.INTERVIEW_TEMPLATES['technical'])
                questions.append({
                    'category': 'Technical',
                    'question': template.format(skill=skill, technology=skill, domain=skill),
                    'rationale': f"Assesses hands-on experience with {skill}, which is mentioned in the job requirements",
                    'difficulty': 'Medium'
                })
        
        # 2. Experience-based Questions
        experience_templates = self.config.INTERVIEW_TEMPLATES['experience']
        actions = ['collaborate', 'solve complex problems', 'meet tight deadlines', 'handle multiple projects']
        
        for i in range(2):
            template = experience_templates[i % len(experience_templates)]
            action = actions[i % len(actions)]
            questions.append({
                'category': 'Experience',
                'question': template.format(action=action, challenge=action, 
                                          team_type='cross-functional', domain='software development'),
                'rationale': f"Evaluates practical experience and problem-solving approach",
                'difficulty': 'Medium'
            })
        
        # 3. Behavioral Questions
        behavioral_questions = self.config.INTERVIEW_TEMPLATES['behavioral']
        for i in range(2):
            question_template = behavioral_questions[i % len(behavioral_questions)]
            questions.append({
                'category': 'Behavioral',
                'question': question_template.format(field='technology'),
                'rationale': "Tests soft skills and cultural fit",
                'difficulty': 'Easy'
            })
        
        # 4. Role-specific Questions
        role_match = self._extract_role_from_job(job_description)
        industry_match = self._extract_industry_from_job(job_description)
        
        role_template = random.choice(self.config.INTERVIEW_TEMPLATES['role_specific'])
        questions.append({
            'category': 'Role-specific',
            'question': role_template.format(role=role_match, industry=industry_match, 
                                           domain=role_match, team=f"{role_match} team"),
            'rationale': f"Assesses genuine interest in the {role_match} role and company fit",
            'difficulty': 'Easy'
        })
        
        # 5. Add scenario-based question if senior role
        if 'senior' in job_lower or 'lead' in job_lower:
            questions.append({
                'category': 'Leadership',
                'question': "Describe a time when you had to make a difficult technical decision that affected the entire team. How did you approach it?",
                'rationale': "Tests leadership and decision-making skills required for senior positions",
                'difficulty': 'Hard'
            })
        
        return questions[:8]  # Return top 8 questions
    
    def _extract_role_from_job(self, job_description: str) -> str:
        """Extract likely role from job description"""
        job_lower = job_description.lower()
        
        roles = {
            'software engineer': ['software engineer', 'developer', 'programmer'],
            'data scientist': ['data scientist', 'data analyst', 'ml engineer'],
            'product manager': ['product manager', 'product owner', 'pm'],
            'designer': ['designer', 'ux designer', 'ui designer'],
            'devops engineer': ['devops', 'site reliability', 'infrastructure'],
            'frontend developer': ['frontend', 'front-end', 'ui developer'],
            'backend developer': ['backend', 'back-end', 'server-side'],
            'full stack developer': ['full stack', 'fullstack', 'full-stack']
        }
        
        for role, keywords in roles.items():
            if any(keyword in job_lower for keyword in keywords):
                return role
        
        return 'software professional'
    
    def _extract_industry_from_job(self, job_description: str) -> str:
        """Extract likely industry from job description"""
        job_lower = job_description.lower()
        
        industries = {
            'fintech': ['fintech', 'financial', 'banking', 'finance'],
            'healthcare': ['healthcare', 'medical', 'health', 'pharma'],
            'ecommerce': ['ecommerce', 'e-commerce', 'retail', 'marketplace'],
            'technology': ['tech', 'software', 'saas', 'platform'],
            'gaming': ['gaming', 'game', 'entertainment'],
            'education': ['education', 'edtech', 'learning'],
            'startup environment': ['startup', 'early stage', 'seed']
        }
        
        for industry, keywords in industries.items():
            if any(keyword in job_lower for keyword in keywords):
                return industry
        
        return 'technology'

print("🤖 FREE AI enhancer ready!")
```

```python
# =============================================================================
# CELL 7: FREE Main Analysis Pipeline
# =============================================================================

class FreeResumeAnalyzer:
    """Complete FREE resume analyzer - NO API keys required!"""
    
    def __init__(self):
        self.doc_processor = DocumentProcessor()
        self.matcher = FreeResumeJobMatcher()
        self.ai_enhancer = FreeAIEnhancer()
        print("🆓 FREE Resume Analyzer initialized successfully!")
    
    def analyze_resume_free(self, file_content: bytes, file_extension: str, 
                           job_description: str) -> Dict:
        """Complete FREE analysis pipeline"""
        
        print("📊 Starting FREE resume analysis...")
        print("=" * 50)
        
        try:
            # Step 1: Extract text from resume
            print("1️⃣ Extracting text from resume...")
            resume_text = self.doc_processor.process_resume(file_content, file_extension)
            
            if not resume_text or len(resume_text.strip()) < 50:
                return {
                    'error': 'Could not extract sufficient text from resume. Please check the file format.',
                    'details': f'Extracted text length: {len(resume_text) if resume_text else 0} characters'
                }
            
            print(f"   ✅ Extracted {len(resume_text)} characters")
            
            # Step 2: Generate comprehensive match report
            print("2️⃣ Analyzing resume-job match (using FREE models)...")
            match_report = self.matcher.generate_comprehensive_report(resume_text, job_description)
            
            print(f"   ✅ Match score: {match_report['match_score']}%")
            print(f"   ✅ Match level: {match_report['match_level']}")
            
            # Step 3: Generate FREE improvement suggestions
            print("3️⃣ Generating improvement suggestions (template-based)...")
            improvements = self.ai_enhancer.generate_free_improvements(
                resume_text, 
                job_description, 
                match_report['skill_gaps'],
                match_report['text_quality_analysis']
            )
            
            print(f"   ✅ Generated {len(improvements)} suggestions")
            
            # Step 4: Generate FREE interview questions
            print("4️⃣ Generating interview questions (template-based)...")
            interview_questions = self.ai_enhancer.generate_free_interview_questions(
                job_description, 
                resume_text,
                match_report['resume_skills']
            )
            
            print(f"   ✅ Generated {len(interview_questions)} questions")
            
            # Step 5: Add recommendations
            print("5️⃣ Generating actionable recommendations...")
            recommendations = self._generate_actionable_recommendations(match_report, improvements)
            
            # Compile final report
            final_report = {
                'status': 'success',
                'match_analysis': match_report,
                'improvement_suggestions': improvements,
                'interview_questions': interview_questions,
                'actionable_recommendations': recommendations,
                'resume_preview': {
                    'first_200_chars': resume_text[:200] + '...',
                    'word_count': len(resume_text.split()),
                    'estimated_reading_time': f"{len(resume_text.split()) // 200 + 1} minutes"
                },
                'analysis_metadata': {
                    'timestamp': pd.Timestamp.now().isoformat(),
                    'processing_method': 'free_models_only',
                    'models_used': [
                        'SentenceTransformers (all-MiniLM-L6-v2)',
                        'spaCy (en_core_web_sm)',
                        'scikit-learn (TF-IDF)',
                        'Template-based AI enhancement'
                    ],
                    'total_skills_identified': sum(len(skills) for skills in match_report['resume_skills'].values() if isinstance(skills, list)),
                    'api_costs': 'FREE - No API calls made'
                }
            }
            
            print("=" * 50)
            print("✅ FREE Analysis complete!")
            print(f"💰 Total cost: $0.00 (completely free!)")
            return final_report
            
        except Exception as e:
            print(f"❌ Analysis failed: {e}")
            return {
                'error': f'Analysis failed: {str(e)}',
                'details': 'Please check your input files and try again.'
            }
    
    def _generate_actionable_recommendations(self, match_report: Dict, improvements: List[Dict]) -> Dict:
        """Generate specific actionable recommendations"""
        
        recommendations = {
            'immediate_actions': [],
            'skill_development': [],
            'resume_optimization': [],
            'interview_preparation': []
        }
        
        match_score = match_report['match_score']
        skill_gaps = match_report['skill_gaps']
        
        # Immediate actions based on match score
        if match_score < 50:
            recommendations['immediate_actions'].extend([
                "Consider applying to similar but slightly different roles to improve match rate",
                "Focus on acquiring 2-3 critical missing skills before applying",
                "Network with professionals in this field to understand requirements better"
            ])
        elif match_score < 75:
            recommendations['immediate_actions'].extend([
                "Apply to this role - you're a decent match with room for improvement",
                "Highlight transferable skills and relevant projects in your application",
                "Prepare strong examples that demonstrate missing skills through related experience"
            ])
        else:
            recommendations['immediate_actions'].extend([
                "Apply immediately - you're a strong match for this role!",
                "Tailor your cover letter to emphasize your matching skills",
                "Prepare for technical interviews focusing on your strong skill areas"
            ])
        
        # Skill development recommendations
        priority_gaps = skill_gaps.get('priority_gaps', {})
        for category, missing_skills in priority_gaps.items():
            if missing_skills:
                recommendations['skill_development'].append(
                    f"Priority learning: {category.replace('_', ' ').title()} - focus on {', '.join(missing_skills[:3])}"
                )
        
        # Resume optimization from improvements
        for improvement in improvements[:3]:
            recommendations['resume_optimization'].append(
                f"{improvement['type']}: {improvement['suggestion'][:100]}..."
            )
        
        # Interview preparation
        strength_areas = skill_gaps.get('strength_areas', {})
        if strength_areas:
            strong_categories = list(strength_areas.keys())[:2]
            recommendations['interview_preparation'].extend([
                f"Prepare detailed examples showcasing your {cat.replace('_', ' ')} experience" 
                for cat in strong_categories
            ])
        
        recommendations['interview_preparation'].append(
            "Practice explaining how you've used similar technologies to demonstrate adaptability"
        )
        
        return recommendations
    
    def create_free_visualization(self, analysis_result: Dict):
        """Create visualizations using matplotlib (FREE)"""
        
        if 'error' in analysis_result:
            print(f"❌ Cannot visualize - Error: {analysis_result['error']}")
            return
        
        match_analysis = analysis_result['match_analysis']
        
        # Create comprehensive visualization
        fig = plt.figure(figsize=(16, 12))
        gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
        
        # 1. Match Score Gauge (Large)
        ax1 = fig.add_subplot(gs[0, :2])
        match_score = match_analysis['match_score']
        colors = ['#ff4444' if match_score < 50 else '#ffa500' if match_score < 75 else '#44ff44']
        
        # Create a half-circle gauge
        angles = np.linspace(0, np.pi, 100)
        scores = np.linspace(0, 100, 100)
        
        # Background gauge
        ax1.fill_between(angles, 0, 1, alpha=0.2, color='gray')
        
        # Score gauge
        score_angle = (match_score / 100) * np.pi
        score_angles = angles[angles <= score_angle]
        if len(score_angles) > 0:
            ax1.fill_between(score_angles, 0, 1, alpha=0.7, color=colors[0])
        
        ax1.set_xlim(0, np.pi)
        ax1.set_ylim(0, 1.2)
        ax1.set_title(f'Overall Match Score: {match_score}%\n({match_analysis["match_level"]} Match)', 
                     fontsize=16, fontweight='bold')
        ax1.axis('off')
        
        # Add score text
        ax1.text(np.pi/2, 0.5, f'{match_score}%', ha='center', va='center', 
                fontsize=24, fontweight='bold', color=colors[0])
        
        # 2. Similarity Metrics Breakdown
        ax2 = fig.add_subplot(gs[0, 2])
        metrics = match_analysis['similarity_metrics']
        metric_names = ['Semantic', 'Keywords', 'TF-IDF']
        metric_values = [
            metrics['semantic_similarity'] * 100,
            metrics['keyword_overlap'] * 100,
            metrics['tfidf_similarity'] * 100
        ]
        
        bars = ax2.bar(metric_names, metric_values, color=['#4285f4', '#ea4335', '#34a853'])
        ax2.set_title('Similarity Breakdown')
        ax2.set_ylabel('Score (%)')
        ax2.set_ylim(0, 100)
        
        # Add value labels on bars
        for bar, value in zip(bars, metric_values):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{value:.1f}%', ha='center', va='bottom', fontsize=10)
        
        # 3. Skill Coverage Heatmap
        ax3 = fig.add_subplot(gs[1, :])
        skill_coverage = match_analysis['skill_gaps']['skill_coverage']
        
        categories = list(skill_coverage.keys())[:6]  # Top 6 categories
        coverages = [skill_coverage[cat] for cat in categories]
        
        # Create heatmap-style visualization
        y_pos = np.arange(len(categories))
        bars = ax3.barh(y_pos, coverages, color=plt.cm.RdYlGn([c/100 for c in coverages]))
        
        ax3.set_yticks(y_pos)
        ax3.set_yticklabels([cat.replace('_', ' ').title() for cat in categories])
        ax3.set_xlabel('Skill Coverage (%)')
        ax3.set_title('Skill Coverage by Category')
        ax3.set_xlim(0, 100)
        
        # Add percentage labels
        for i, (bar, coverage) in enumerate(zip(bars, coverages)):
            width = bar.get_width()
            ax3.text(width + 2, bar.get_y() + bar.get_height()/2,
                    f'{coverage:.1f}%', ha='left', va='center', fontsize=10)
        
        # 4. Skills Gap Analysis
        ax4 = fig.add_subplot(gs[2, 0])
        missing_counts = []
        matching_counts = []
        gap_categories = []
        
        for category in list(skill_coverage.keys())[:5]:
            missing = len(match_analysis['skill_gaps']['missing_skills'].get(category, []))
            matching = len(match_analysis['skill_gaps']['matching_skills'].get(category, []))
            
            if missing > 0 or matching > 0:  # Only show categories with data
                gap_categories.append(category.replace('_', ' ')[:10])  # Truncate long names
                missing_counts.append(missing)
                matching_counts.append(matching)
        
        if gap_categories:
            x = np.arange(len(gap_categories))
            width = 0.35
            
            ax4.bar(x - width/2, matching_counts, width, label='Have', color='#44ff44', alpha=0.7)
            ax4.bar(x + width/2, missing_counts, width, label='Missing', color='#ff4444', alpha=0.7)
            
            ax4.set_title('Skills: Have vs Missing')
            ax4.set_xlabel('Categories')
            ax4.set_ylabel('Number of Skills')
            ax4.set_xticks(x)
            ax4.set_xticklabels(gap_categories, rotation=45, ha='right')
            ax4.legend()
        
        # 5. Improvement Priority
        ax5 = fig.add_subplot(gs[2, 1])
        improvements = analysis_result['improvement_suggestions']
        improvement_types = {}
        priorities = {'High': 0, 'Medium': 0, 'Low': 0}
        
        for improvement in improvements:
            imp_type = improvement.get('type', 'General')
            priority = improvement.get('priority', 'Medium')
            improvement_types[imp_type] = improvement_types.get(imp_type, 0) + 1
            priorities[priority] += 1
        
        if priorities and sum(priorities.values()) > 0:
            labels = list(priorities.keys())
            sizes = list(priorities.values())
            colors_pie = ['#ff4444', '#ffa500', '#44ff44']
            
            ax5.pie(sizes, labels=labels, colors=colors_pie, autopct='%1.0f%%', startangle=90)
            ax5.set_title('Improvement Priority')
        
        # 6. Text Quality Analysis
        ax6 = fig.add_subplot(gs[2, 2])
        quality = match_analysis['text_quality_analysis']
        
        quality_metrics = {
            'Word Count': min(quality.get('word_count', 0) / 10, 100),  # Scale to 100
            'Has Metrics': 100 if quality.get('has_quantifiable_achievements', False) else 0,
            'Action Verbs': 100 if quality.get('has_action_verbs', False) else 0,
            'Quality Score': quality.get('quality_score', 0)
        }
        
        angles_radar = np.linspace(0, 2*np.pi, len(quality_metrics), endpoint=False).tolist()
        values = list(quality_metrics.values())
        angles_radar += angles_radar[:1]  # Complete the circle
        values += values[:1]
        
        ax6 = plt.subplot(gs[2, 2], projection='polar')
        ax6.plot(angles_radar, values, 'o-', linewidth=2, color='#4285f4')
        ax6.fill(angles_radar, values, alpha=0.25, color='#4285f4')
        ax6.set_xticks(angles_radar[:-1])
        ax6.set_xticklabels(quality_metrics.keys(), fontsize=8)
        ax6.set_ylim(0, 100)
        ax6.set_title('Resume Quality', y=1.1)
        
        plt.suptitle('FREE AI Resume Analysis Report', fontsize=18, fontweight='bold', y=0.95)
        plt.tight_layout()
        plt.show()
        
        # Print summary
        print("\n" + "="*60)
        print("📊 ANALYSIS SUMMARY")
        print("="*60)
        print(f"💯 Overall Match Score: {match_score}% ({match_analysis['match_level']} Match)")
        print(f"🎯 Skills Matched: {match_analysis['skill_gaps']['total_skills_matched']}")
        print(f"📈 Top Strength: {max(skill_coverage.items(), key=lambda x: x[1])[0].replace('_', ' ').title()}")
        print(f"🔧 Priority Gap: {list(match_analysis['skill_gaps'].get('priority_gaps', {}).keys())[0].replace('_', ' ').title() if match_analysis['skill_gaps'].get('priority_gaps') else 'None'}")
        print(f"💡 Improvement Areas: {len(improvements)} identified")
        print(f"❓ Interview Questions: {len(analysis_result['interview_questions'])} prepared")
        print("💰 Total Cost: FREE!")
        print("="*60)

print("📈 FREE visualization system ready!")
```

```python
# =============================================================================
# CELL 8: FREE FastAPI Setup (No API Keys Required)
# =============================================================================

# Create FastAPI app
free_app = FastAPI(
    title="FREE AI Resume Analyzer", 
    version="1.0.0",
    description="Completely FREE resume analysis using open-source models only!"
)

# Add CORS
free_app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Initialize FREE analyzer
free_analyzer = FreeResumeAnalyzer()

@free_app.post("/analyze-free")
async def analyze_resume_free_endpoint(
    resume_file: UploadFile = File(...),
    job_description: str = Form(...)
):
    """FREE resume analysis endpoint - NO API keys required!"""
    try:
        # Validate file size (limit to 10MB)
        file_content = await resume_file.read()
        if len(file_content) > 10 * 1024 * 1024:  # 10MB limit
            return {
                "status": "error",
                "message": "File too large. Please upload a file smaller than 10MB."
            }
        
        # Get file extension
        file_extension = os.path.splitext(resume_file.filename)[1]
        
        if not file_extension:
            return {
                "status": "error", 
                "message": "Could not determine file type. Please ensure your file has a proper extension (.pdf, .docx, .txt)"
            }
        
        # Validate job description
        if not job_description or len(job_description.strip()) < 50:
            return {
                "status": "error",
                "message": "Please provide a more detailed job description (at least 50 characters)"
            }
        
        # Perform FREE analysis
        result = free_analyzer.analyze_resume_free(file_content, file_extension, job_description)
        
        return result
        
    except Exception as e:
        return {
            "status": "error",
            "message": f"Analysis failed: {str(e)}",
            "details": "Please check your inputs and try again."
        }

@free_app.get("/health-free")
async def health_check_free():
    """Health check for FREE service"""
    return {
        "status": "healthy", 
        "message": "FREE Resume Analyzer API is running",
        "features": [
            "Document parsing (PDF, DOCX, TXT)",
            "Skill extraction and matching",
            "Resume-job similarity scoring",
            "Improvement suggestions",
            "Mock interview questions",
            "All completely FREE!"
        ],
        "models_loaded": {
            "sentence_transformer": "✅ all-MiniLM-L6-v2",
            "spacy_nlp": "✅ en_core_web_sm",
            "sklearn_vectorizers": "✅ TF-IDF & Count",
            "document_parsers": "✅ PyPDF2, python-docx, pdfplumber"
        }
    }

@free_app.get("/")
async def root():
    """Root endpoint with API information"""
    return {
        "message": "Welcome to FREE AI Resume Analyzer!",
        "description": "Complete resume analysis using only free, open-source models",
        "cost": "FREE - No API keys required!",
        "endpoints": {
            "POST /analyze-free": "Upload resume and job description for analysis",
            "GET /health-free": "Check service health and loaded models",
            "GET /docs": "Interactive API documentation"
        },
        "supported_formats": ["PDF", "DOCX", "TXT"],
        "features": [
            "Resume parsing and text extraction",
            "Skills gap analysis", 
            "Match score calculation",
            "Improvement suggestions",
            "Mock interview questions",
            "Quality analysis"
        ]
    }

print("🚀 FREE FastAPI backend ready!")
```
